```python
def get_nth_derivative(func_expr, var_name, n, c):
    x = sp.Symbol(var_name)
    expr = sp.sympify(func_expr)
    derivative_expr = sp.diff(expr, x, n)
    result = derivative_expr.subs(x, c)
    return result.evalf()

```

a

```python
crit_a = [-1, 2]
for crit_val in crit_a:
    for i in range(1, 11):
        val = get_nth_derivative("((x + 1) ** 3) / (x ** 2)", "x", i, crit_val)
        print(f"The derivative of order {i} at {crit_val} is {val}.")
    print()
```
