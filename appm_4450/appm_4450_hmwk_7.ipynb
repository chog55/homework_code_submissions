```python
import numpy as np, matplotlib.pyplot as plt
```

```python
g = lambda n, x: x ** n / n
```

```python
plt.figure()
x = np.linspace(0, 1, 100)
for i in range(1, 50, 3):
    plt.plot(x, g(i, x), label = rf"$g_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$g_n(x)=\frac{x^n}{n}$")
plt.savefig('figures/1a_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
g_p = lambda n, x: x ** (n - 1)
```

```python
plt.figure()
x = np.linspace(0, 1, 100)
for i in range(1, 50, 3):
    plt.plot(x, g_p(i, x), label = rf"$g'_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$g'_n(x)=\frac{x^n}{n}$")
plt.savefig('figures/1b_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
h = lambda n, x: (np.sin(n * x)) / (np.sqrt(x))
x = np.linspace(-5, 5, 500)
for i in range(1, 50, 3):
    plt.plot(x, h(i, x), label = rf"$h_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$h_n(x)=\frac{\sin(nx)}{\sqrt{n}}$")
plt.savefig('figures/2_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```

```python
h_p = lambda n, x: np.sqrt(n) * np.cos(n * x)
x = np.linspace(-5, 5, 500)
for i in range(1, 50, 3):
    plt.plot(x, h_p(i, x), label = rf"$h'_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$h'_n(x)=\sqrt{n}\cos(nx)$")
plt.savefig('figures/2_2.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```

```python
f = lambda n, x: (1 - np.abs(x) ** n) / (1 + np.abs(x) ** n)
x = np.linspace(-5, 5, 500)
for i in range(1, 50, 3):
    plt.plot(x, f(i, x), label = rf"$f_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$f_n(x)=\frac{1-|x|^n}{1+|x|^n}$")
plt.savefig('figures/3a_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```

```python
f = lambda n, x: 1 / (1 + x ** n)
x = np.linspace(2, 10, 500)
for i in range(1, 50, 3):
    plt.plot(x, f(i, x), label = rf"$f_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$f_n(x)=\frac{1}{1+x^n}$")
plt.savefig('figures/3b_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```

```python
f = lambda n, x: 1 / (n * x + 1)
x = np.linspace(0, 1, 500)
for i in range(1, 50, 3):
    plt.plot(x, f(i, x), label = rf"$f_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$f_n(x)=\frac{1}{nx+1}$")
plt.savefig('figures/3c_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```

```python
f = lambda n, x: x / (n * x + 1)
x = np.linspace(0, 1, 500)
for i in range(1, 50, 3):
    plt.plot(x, f(i, x), label = rf"$f_{{{i}}}$")
plt.legend(bbox_to_anchor = (1, 0.5), loc = 'center left')
plt.title(r"$f_n(x)=\frac{x}{nx+1}$")
plt.savefig('figures/3d_1.png', dpi = 300, bbox_inches = 'tight')
plt.show()  
```
