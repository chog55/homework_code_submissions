```python
import matplotlib.pyplot as plt, numpy as np, random, sympy as sp, math
```

3b

```python
def E(n):
    count = 0
    for i in range(n):
        x_val = random.random()
        y_val = random.random()
        if y_val < 1 / (1 + x_val ** 6):
            count += 1
    return count / n
```

```python
x = np.arange(1, 5.1, 0.1)
N = np.int64(10 ** x)
plt.semilogx(N, [E(n) for n in N], label = "E(N)")
plt.xlabel("N")
plt.axhline(0.904, color = "orange", label = "0.904: Value of integral obtained by\nnumerical integration on HP Prime calculator")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5))
```

```python
res = []
for i in range(10 ** 5):
    t = 0
    for j in range(3):
        x = random.random()
        t = max(t, x)
    res.append(t)
plt.hist(res, density = True, bins = 40, ec = "black", label = r"$\max_{i=1,2,3} \{X_i\}$")
t = np.linspace(0, 1, 500)
plt.plot(t, 3 * t ** 2, label = r"$f(t)=3t^2$")
plt.xlabel("Hours elapsed since 6PM (t)")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5))
```
