```python
q = 0.4
p = 0.35
s = 0.25
total = 0
n = 100000
for _ in range(n):
    i = 10
    while i > 0:
        u = random.random()
        if u <= q:
            i -= 1
        elif u <= q + p:
            i += 1
        else:
            break
    total += i
print(total / n)
```

```python
e = lambda p, q, s, i: i + ((p - q) / s) * (1 - ((1 - math.sqrt(1 - 4 * p * q)) / (2 * p)) ** i)
print(e(p, q, s, 10))
```

```python
p = np.zeros((6,6))
for i in range(6):
    for j in range(6):
        if i == 0 and j == 5:
            p[i,j] = 1
        elif i != 0 and i >= j:
            p[i,j] = math.comb(i, j) * 0.1 ** (i - j) * 0.9 ** j
        else:
            p[i,j] = 0
print(p)
for i in range(6):
    print(np.sum(p[i,:]))
```
