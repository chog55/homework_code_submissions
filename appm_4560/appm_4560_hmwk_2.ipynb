```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

1

```python
def f(x, gamma, x_0):
    return np.where(x >= x_0, (gamma - 1) * x_0 ** (gamma - 1) * x ** (-gamma), 0.0)
```

```python
def f_inv(u, gamma, x_0):
    return x_0 * (1 - u) ** (-1 / (gamma - 1))
```

```python
res = []
for i in range(100):
    u = random.random()
    res.append(f_inv(u, 4, 10))
bins = np.linspace(10, 60, 46)
plt.hist(res, density = True, bins = bins, ec = "black", range=(0, 60), label = r"$u$")
x = np.linspace(0, 60, 500)
plt.plot(x, f(x, 4, 10), label = r"$f(x)$")
plt.xlabel("x")
plt.xlim(0, 60)
plt.title("N = 100")
plt.legend(loc = "center left", bbox_to_anchor = (1, 0.5))
```
