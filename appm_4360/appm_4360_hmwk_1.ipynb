```python
import numpy as np, matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 4, 500)
b = np.linspace(-2, 2, 500)
re, im = np.meshgrid(a, b)
def f(A, B):
    a = 2 * A - 4
    b = 2 * B
    return np.sqrt(a ** 2 + b ** 2)
g = ((f(re, im) > 1) & (f(re, im) < 2)).astype(float)
plt.pcolormesh(re, im, g)
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.colorbar()
```
